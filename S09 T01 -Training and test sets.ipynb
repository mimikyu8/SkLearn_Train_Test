Parteix el conjunt de dades DelayedFlights.csv en train i test. Estudia els dos conjunts per separat, a nivell descriptiu.

Split the DelayedFlights.csv dataset into train and test. Study both sets separately, at descriptive level.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DelayedFlights.csv", index_col = 0)
```

The data consists of flight arrival and departure details for all commercial flights within the USA. Variable descriptions:
* 1 Year; 2008 
* 2 Month; 1-12 
* 3 DayofMonth; 1-31 
* 4 DayOfWeek; 1 (Monday) - 7 (Sunday) 
* 5 DepTime; actual departure time (local, hhmm) 
* 6 CRSDepTime; scheduled departure time (local, hhmm) 
* 7 ArrTime; actual arrival time (local, hhmm) 
* 8 CRSArrTime; scheduled arrival time (local, hhmm) 
* 9 UniqueCarrier; unique carrier code 
* 10 FlightNum; flight number 
* 11 TailNum; plane tail number 
* 12 ActualElapsedTime; in minutes 
* 13 CRSElapsedTime; in minutes 
* 14 AirTime; in minutes 
* 15 ArrDelay; arrival delay, in minutes 
* 16 DepDelay; departure delay, in minutes 
* 17 Origin; origin IATA airport code 
* 18 Dest; destination IATA airport code 
* 19 Distance; in miles 
* 20 TaxiIn; taxi in time, in minutes 
* 21 TaxiOut; taxi out time in minutes 
* 22 Cancelled; was the flight cancelled? 
* 23 CancellationCode; reason for cancellation (A = carrier, B = weather, C = NAS, D = security) 
* 24 Diverted; 1 = yes, 0 = no 
* 25 CarrierDelay; in minutes 
* 26 WeatherDelay; in minutes 
* 27 NASDelay; in minutes 
* 28 SecurityDelay; in minutes 
* 29 LateAircraftDelay; in minutes

```python
df["Year"].unique()
```

```python
#drop the year column because it always says 2008
df = df.drop(["Year"], axis = 1)
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size = 0.3, random_state = 42)
```

```python
print("Train shape:", train.shape, "\nTest shape:", test.shape)
```

```python
print("ActualElapsedTime train mean:", train["ActualElapsedTime"].mean())
print("ActualElapsedTime train standard deviation:", train["ActualElapsedTime"].std())
print("ActualElapsedTime train median:", train["ActualElapsedTime"].median())
print("ActualElapsedTime train mode:", train["ActualElapsedTime"].mode())
print("In minutes\n")


print("ActualElapsedTime test mean:", test["ActualElapsedTime"].mean())
print("ActualElapsedTime test standard deviation:", test["ActualElapsedTime"].std())
print("ActualElapsedTime test median:", test["ActualElapsedTime"].median())
print("ActualElapsedTime test mode:", test["ActualElapsedTime"].mode())
print("In minutes\n")


print("CRSElapsedTime train mean:", train["CRSElapsedTime"].mean())
print("CRSElapsedTime train standard deviation:", train["CRSElapsedTime"].std())
print("CRSElapsedTime train median:", train["CRSElapsedTime"].median())
print("CRSElapsedTime train mode:", train["CRSElapsedTime"].mode())
print("In minutes\n")


print("CRSElapsedTime test mean:", test["CRSElapsedTime"].mean())
print("CRSElapsedTime test standard deviation:", test["CRSElapsedTime"].std())
print("CRSElapsedTime test median:", test["CRSElapsedTime"].median())
print("CRSElapsedTime test mode:", test["CRSElapsedTime"].mode())
print("In minutes\n")


print("AirTime train mean:", train["AirTime"].mean())
print("AirTime train standard deviation:", train["AirTime"].std())
print("AirTime train median:", train["AirTime"].median())
print("AirTime train mode:", train["AirTime"].mode())
print("In minutes\n")

print("AirTime test mean:", test["AirTime"].mean())
print("AirTime test standard deviation:", test["AirTime"].std())
print("AirTime test median:", test["AirTime"].median())
print("AirTime test mode:", test["AirTime"].mode())
print("In minutes\n")


print("ArrDelay train mean:", train["ArrDelay"].mean())
print("ArrDelay train standard deviation:", train["ArrDelay"].std())
print("ArrDelay train median:", train["ArrDelay"].median())
print("ArrDelay train mode:", train["ArrDelay"].mode())
print("In minutes\n")


print("ArrDelay test mean:", test["ArrDelay"].mean())
print("ArrDelay test standard deviation:", test["ArrDelay"].std())
print("ArrDelay test median:", test["ArrDelay"].median())
print("ArrDelay test mode:", test["ArrDelay"].mode())
print("In minutes\n")


print("DepDelay train mean:", train["DepDelay"].mean())
print("DepDelay train standard deviation:", train["DepDelay"].std())
print("DepDelay train median:", train["DepDelay"].median())
print("DepDelay train mode:", train["DepDelay"].mode())
print("In minutes\n")


print("DepDelay test mean:", train["DepDelay"].mean())
print("DepDelay test standard deviation:", train["DepDelay"].std())
print("DepDelay test median:", train["DepDelay"].median())
print("DepDelay test mode:", train["DepDelay"].mode())
print("In minutes\n")


print("Distance test mean:", test["Distance"].mean())
print("Distance test standard deviation:", test["Distance"].std())
print("Distance test median:", test["Distance"].median())
print("Distance test mode:", test["Distance"].mode())
print("In miles\n")


print("Distance train mean:", train["Distance"].mean())
print("Distance train standard deviation:", train["Distance"].std())
print("Distance train median:", train["Distance"].median())
print("Distance train mode:", train["Distance"].mode())
print("In miles")
```

As we can see, the summary statistics of each of these selected are significantly simmilar.  We can assume that since both sets are such big samples of the main dataset, they are an accurate representation of it and of eachother. 

```python
sns.countplot( x = "UniqueCarrier", data = train, color = "#457b9d", order = train["UniqueCarrier"].value_counts().index).set(
    title = "Train - Number of flights per unique carrier")
```

```python
sns.countplot( x = "UniqueCarrier", data = test, color = "#1d3557", order = test["UniqueCarrier"].value_counts().index).set(
    title = "Test - Number of flights per unique carrier")
```

As we can see in the above graphs, the distribution of the number of flights per carrier is very similar. We can assume that since both sets are such big samples of the main dataset, they are an accurate representation of it.
