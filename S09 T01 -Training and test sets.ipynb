Split the DelayedFlights.csv dataset into train and test. Study both sets separately, at descriptive level.

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("DelayedFlights.csv", index_col = 0)
```

The data consists of flight arrival and departure details for all commercial flights within the USA. Variable descriptions:
* 1 Year; 2008 
* 2 Month; 1-12 
* 3 DayofMonth; 1-31 
* 4 DayOfWeek; 1 (Monday) - 7 (Sunday) 
* 5 DepTime; actual departure time (local, hhmm) 
* 6 CRSDepTime; scheduled departure time (local, hhmm) 
* 7 ArrTime; actual arrival time (local, hhmm) 
* 8 CRSArrTime; scheduled arrival time (local, hhmm) 
* 9 UniqueCarrier; unique carrier code 
* 10 FlightNum; flight number 
* 11 TailNum; plane tail number 
* 12 ActualElapsedTime; in minutes 
* 13 CRSElapsedTime; in minutes 
* 14 AirTime; in minutes 
* 15 ArrDelay; arrival delay, in minutes 
* 16 DepDelay; departure delay, in minutes 
* 17 Origin; origin IATA airport code 
* 18 Dest; destination IATA airport code 
* 19 Distance; in miles 
* 20 TaxiIn; taxi in time, in minutes 
* 21 TaxiOut; taxi out time in minutes 
* 22 Cancelled; was the flight cancelled? 
* 23 CancellationCode; reason for cancellation (A = carrier, B = weather, C = NAS, D = security) 
* 24 Diverted; 1 = yes, 0 = no 
* 25 CarrierDelay; in minutes 
* 26 WeatherDelay; in minutes 
* 27 NASDelay; in minutes 
* 28 SecurityDelay; in minutes 
* 29 LateAircraftDelay; in minutes

# Exercici 2

Aplica algun procés de transformació (estandarditzar les dades numèriques, crear columnes dummies, polinomis...).

Apply some transformation process (standarize numerical values, create dummy columns, polynomials...).

```python
df.head()
```

We are first deleting some columns that are not needed. The Year column, that is always 2008; the Cancelled column, as well as the rows that have cancelled = 1, because those are of no interest if we are studying delay; the CancellationCode column, for the same reasons; the Diverted column, as well as the rows that have diverted = 1, because those are of no interest if we are studying delay; and the Origin and Dest because they add no real information that the column Distance doesn't provide, as well as the TailNum column.

```python
#shows Year column only has one value
df["Year"].unique()
```

```python
#remove the Year column
df = df.drop(["Year"], axis = 1)
```

```python
#keep only rows that were not cancelled
df = df[df.Cancelled == 0]

#remove the Cancelled row
df = df.drop(["Cancelled"], axis = 1)
```

```python
#remove CancellationCode column
df = df.drop(["CancellationCode"], axis = 1)
```

```python
#keep only rows that were not diverted
df = df[df.Diverted == 0]

#remove the Diverted row
df = df.drop(["Diverted"], axis = 1)
```

```python
#remove the Origin and Dest row
df = df.drop(["Origin", "Dest", "TailNum"], axis = 1)
```

```python
df.head()
```

Now we will create some dummy variables for the UniqueCarrier column.

```python
#create dummys for the atribute UniqueCarrier
dummy = pd.get_dummies(df.UniqueCarrier)

dummy.head()
```

Next, we will normalize the numerical values.

```python
plt.hist(df.DepTime, edgecolor = "black", linewidth = 1, color = "#5FAD56")
plt.title("Actual departure time")
```

```python
plt.hist(df.CRSDepTime, edgecolor = "black", linewidth = 1, color = "#F2C14E")
plt.title("Scheduled departure time")
```

As the above plots show, the scheduled and real departure times don't have a normal distribution, so we will have to normalize these values. We can expect that the scheduled and real arrival times have as well a non normal distribution. 

```python
plt.hist(df.ArrTime, edgecolor = "black", linewidth = 1, color = "#F78154")
plt.title("Actual arrival time")
```

```python
plt.hist(df.CRSArrTime, edgecolor = "black", linewidth = 1, color = "#4D9078")
plt.title("Scheduled arrival time")
```

The scheduled and real arrival times have as well a non gaussian distribution and hence will be normalized. 

```python
from sklearn.preprocessing import MinMaxScaler

c = df[["DayofMonth", "DayOfWeek", "DepTime", "CRSDepTime", "ArrTime", "CRSArrTime", "AirTime", "ArrDelay", "DepDelay",
        "Distance", "TaxiIn", "TaxiOut", "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"]]

#standarizing numeric atributes
mms = MinMaxScaler().fit_transform(c)
```

```python
mms_df = pd.DataFrame(data = mms, columns = ["DayofMonth", "DayOfWeek", "DepTime", "CRSDepTime", "ArrTime", "CRSArrTime",
                                              "AirTime", "ArrDelay", "DepDelay", "Distance", "TaxiIn", "TaxiOut", 
                                              "CarrierDelay", "WeatherDelay", "NASDelay", "SecurityDelay", "LateAircraftDelay"])
mms_df.head()
```

```python
#create transformed dataset
final_df = pd.concat([mms_df, dummy], axis = 1)
final_df.head()
```

We can see that through the process of normalization, each column got a minimum value of 0 and a maximum value of 1. However, they still have different quartiles, mean and standard deviation as the values still have the same distribution as before.

The above graphs show the same data that was plotted before normalization, and in this case we can see that the values always fall between 0 and 1.
